```python
import numpy as np
```

```python
arr = np.arange(1, 21)
print("Array:", arr)

# Statistics
print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Standard Deviation:", np.std(arr))

# Indices where elements > 10
indices_gt_10 = np.where(arr > 10)
print("Indices of elements > 10:", indices_gt_10)
```

```python
arr_2d = np.arange(1, 17).reshape(4, 4)
print("Original Array:\n", arr_2d)
print("Transpose:\n", arr_2d.T)
print("Row-wise Sum:", np.sum(arr_2d, axis=1))
print("Column-wise Sum:", np.sum(arr_2d, axis=0))
```

```python
a = np.random.randint(1, 21, (3, 3))
b = np.random.randint(1, 21, (3, 3))

print("Array A:\n", a)
print("Array B:\n", b)
print("Addition:\n", a + b)
print("Subtraction:\n", a - b)
print("Multiplication:\n", a * b)
print("Dot Product:\n", np.dot(a, b))
```

```python
import pandas as pd
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)
print(df.head())
```

```python
print("Summary (Age & Salary):\n", df[['Age', 'Salary']].describe())

# Average salary in HR
avg_hr_salary = df[df['Department'] == 'HR']['Salary'].mean()
print("Average HR Salary:", avg_hr_salary)
```

```python
df['Bonus'] = df['Salary'] * 0.10
print(df)
```

```python
filtered = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print("Employees aged between 25 and 30:\n", filtered)
```

```python
dept_avg = df.groupby('Department')['Salary'].mean()
print("Average Salary by Department:\n", dept_avg)
```

```python
sorted_df = df.sort_values(by='Salary')
sorted_df.to_csv("sorted_employees.csv", index=False)
print("Sorted DataFrame saved to 'sorted_employees.csv'")
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.plot(x, y, marker='o')
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.bar(students, marks, color='skyblue')
plt.title("Marks by Student")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

explode = [0.1, 0, 0, 0]  # Highlight North America
plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%', shadow=True)
plt.title("Revenue Distribution")
plt.show()
```
